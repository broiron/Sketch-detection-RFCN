```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
img = cv2.imread('./predict/results/Line_Dataset-100_res.jpg')

plt.imshow(img)
```

```python
plt.figure(figsize=(7, 10))

for i in range(20):

# 열개수 4개, 행 개수 5개 짜리 그리드를 생성한 후, 인덱스 i+1 번째를 사용
    plt.subplot(5, 4, i + 1)

# 빈 이미지에 인덱스 숫자를 출력
img_empty = np.zeros((150,150,1), dtype=np.uint8)
cv2.putText(img_empty, str(i+1), (30,90), cv2.FONT_HERSHEY_PLAIN, 5, (255,255,255), 2, cv2.LINE_AA)

# img_empty=img_empty /255

# 이미지를 subplot에서 지정한 인덱스에 보여줌
plt.imshow(img_empty)

#subplot 위에 인덱스 번호 출력
plt.title('index=' + str(i+1), color='blue', fontsize=15)
plt.axis('off')

# 전체 제목을 출력
plt.suptitle('Number', fontsize=25)
plt.tight_layout() # subplot간 여백을 자동으로 조정
plt.show()
```
